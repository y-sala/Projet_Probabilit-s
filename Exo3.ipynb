### Compter le nombre de un, et, le, il, est

```python
def compter_mot(mot_cherche):
    mots=[]
    a=0
    for i in range(len(mot)):
        if(mot[i]==mot_cherche):
            a+=1
            mots.append(a)
        else:
            mots.append(0)
    return a, mots
```
