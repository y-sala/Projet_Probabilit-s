```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Simulation sous python

```python
def realize(n):
    '''
    On initialise le tableau qui contient les différentes réalisations {y1, y2, ..., yn} des n variables {Y1, Y2, ..., Yn}
    On distribue selon une loi uniforme chaque variable Yi et on le met dans le tableau y
    '''
    y=[]
    for i in range(n):
        y.append(np.random.uniform(0, 10))
    return y
```

### 2. Réalisation moyenne empirique

```python
def moy_Emp(n):
    '''
    On calcule la moyenne empirique pour chaque réalisation avec un échantillon n
    '''
    return (n**-1)*np.sum(realize(n))
```

### 3. Répeter 5000 fois les étapes 1 et 2

```python
def repeter(N,n):
    '''
    On répète N fois les fonctions realize & moy_Emp
    '''
    ybarre=[]
    for i in range(N):
        Y=moy_Emp(n)
        ybarre.append(Y)
    return ybarre
```

### 4. On construit l'histogramme des 5000 réalisations

```python
def hist(N, n):
    v=repeter(N, n)
    plt.hist(v, color='black')
    plt.grid()
    plt.title('Lorsque N='+str(N)+' et n='+str(n))
    plt.show()
hist(N=5000, n=10)
```

### 5. Répeter l'expérience pour n=10, 100, 1000, 10000 et N=10000

```python
def exp_mult(N, n):
    for i in range(len(n)):
        hist(N, n[i])
exp_mult(10000, [10, 100, 1000, 10000])
```
